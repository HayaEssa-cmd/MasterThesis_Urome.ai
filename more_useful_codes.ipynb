<strong><h1>Finding all AR genes</h1></strong>

```python
import pandas as pd

# Define the Excel file path
excel_file_path = r'C:\Users\hayae\Desktop\New_opencravat_summary_sheets.xlsx'

# Define a list of sheet names to read
sheets_to_read = ['PR_001_new', 'PR_014_new', 'PR_015_new', 'PR_016_new', 'PR_017_new', 'PR_018_new', 
                  'PR_020_new', 'PR_021_new', 'PR_122_new', 'PR_126_new', 'PR_127_new', 'PR_128_new', 
                  'PR_133_new', 'PR_135_new', 'PR_138_new', 'PR_142_new', 'PR_175_new', 'PR_189_new', 
                  'PR_200_new', 'PR_201_new', 'PR_209_new', 'PR_245_new', 'PR_270_new', 'PR_282_new', 
                  'PR_283_new', 'PR_298_1_new', 'PR_28', 'PR_35', 'PR_36', 'PR_37', 'PR_38', 
                  'PR_39', 'PR_40', 'PR_41', 'PR_42', 'PR_43', 'PR_44', 'PR_45', 'PR_46', 'PR_48', 
                  'PR_49', 'PR_50', 'PR_52', 'PR_53', 'PR_54', 'PR_55', 'PR_56', 'PR_57', 
                  'PR_58', 'PR_59', 'PR_068', 'PR_070', 'PR_080', 'PR_081', 'PR_094', 'PR_095', 
                  'PR_096', 'PR_001_v2', 'PR_015_v2', 'PR_023_v2', 'PR_064_v2', 'PR_066_v2', 'PR_067_v2',
                  'PR_068_v1', 'PR_071', 'PR_072', 'PR_077', 'PR_080_v1', 'PR_081_v1', 'PR_109', 'PR_110', 'PR_112',
                  'PR_113', 'PR_115', 'PR_117', 'PR_119', 'PR_120', 'PR_121', 'PR_123', 'PR_124', 'PR_125', 'PR_126', 
                  'PR_127', 'PR_128']

# Genes to search for
genes_to_search = ["ARID1A", "ATM", "ATR", "ATRX", "BARD1", "BRCA1", "BRCA2", "BRIP1", "CDK12", 
                   "CHEK1", "CHEK2", "FANCA", "FANCF", "FANCG", "FANCI", "FANCL", "FANCM", 
                   "HDAC2", "MLH1", "MRE11A", "MSH2", "NBN", "PALB2", "PALP2", "PPP2R2A", "RAD50", 
                   "RAD51", "RAD51B", "RAD51C", "RAD51D", "RAD54", "RAD54L", "WRN"]

# Variants to exclude
variants_to_exclude = [
    {"Gene": "ATRX", "cDNA_change": "c.2785G>C"},
    {"Gene": "BARD1", "cDNA_change": "c.1518_1519delinsCA"},
    {"Gene": "FANCA", "cDNA_change": "c.2574C>G"}
]

# Create an empty DataFrame to hold the final results
final_result = pd.DataFrame()

# Iterate over each sheet to read
for sheet_name in sheets_to_read:
    # Read the Excel file
    df = pd.read_excel(excel_file_path, sheet_name=sheet_name)
    
    # Apply the initial filters
    df_filtered = df[(df['Alternate_reads'] >= 5) & (df['Total_reads'] >= 500) & (df['Variant_AF'] >= 0.005)]
    
    # Exclude specific variants
    for variant in variants_to_exclude:
        df_filtered = df_filtered[~((df_filtered['Gene'] == variant['Gene']) & (df_filtered['cDNA_change'] == variant['cDNA_change']))]
    
    # Create a dictionary to hold gene presence (1) or absence (0)
    gene_presence = {}
    
    # Iterate over genes to search for
    for gene in genes_to_search:
        # Check if the gene exists in the 'Gene' column
        if gene in df_filtered['Gene'].values:
            gene_presence[gene] = 1
        else:
            gene_presence[gene] = 0
            
    # Convert dictionary to DataFrame
    result_df = pd.DataFrame(gene_presence.items(), columns=['Gene', sheet_name])
    
    # Set the Gene column as the index
    result_df.set_index('Gene', inplace=True)
    
    # Transpose the DataFrame to have genes as columns and sheet names as rows
    result_df = result_df.T
    
    # Append the transposed result DataFrame to the final result DataFrame
    final_result = pd.concat([final_result, result_df])

# Reset the index to include the column with sheet names
final_result.reset_index(inplace=True)
final_result.rename(columns={'index': 'Sheet'}, inplace=True)

# Print the final result
print(final_result)

# Save the final result to a new Excel file
final_result.to_excel(r'C:\Users\hayae\Desktop\AR_gene_presence_all_sheets.xlsx', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a DataFrame
df = pd.read_excel("C:/Users/hayae/Desktop/Onco_test.xlsx", sheet_name="AR")

# Ensure both 'Yes' and 'No' categories are present even if all entries are 'Yes'
high_low_volume_counts = df['high volume, low volume'].value_counts().reindex(['Yes', 'No'], fill_value=0)
high_low_risk_counts = df['high risk, low risk'].value_counts().reindex(['Yes', 'No'], fill_value=0)

# Calculate percentages
high_low_volume_percentages = high_low_volume_counts / high_low_volume_counts.sum() * 100
high_low_risk_percentages = high_low_risk_counts / high_low_risk_counts.sum() * 100

# Create a bar plot
plt.figure(figsize=(16, 10))

# Plot for 'high volume, low volume'
plt.subplot(1, 2, 1)
high_low_volume_counts.plot(kind='bar', color='#E69F00')
for index, value in enumerate(high_low_volume_counts):
    if high_low_volume_percentages[index] == 0:
        plt.text(index, value + 0.5, f'n={value} ({high_low_volume_percentages[index]:.1f}%)', ha='center', fontsize=12)
    else:
        plt.text(index, value - 0.5, f'n={value} ({high_low_volume_percentages[index]:.1f}%)', ha='center', fontsize=12)
plt.title('High Volume vs. Low Volume', fontsize=18)
plt.xlabel('Volume', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)

# Plot for 'high risk, low risk'
plt.subplot(1, 2, 2)
high_low_risk_counts.plot(kind='bar', color='#56B4E9')
for index, value in enumerate(high_low_risk_counts):
    if high_low_risk_percentages[index] == 0:
        plt.text(index, value + 0.5, f'n={value} ({high_low_risk_percentages[index]:.1f}%)', ha='center', fontsize=12)
    else:
        plt.text(index, value - 0.5, f'n={value} ({high_low_risk_percentages[index]:.1f}%)', ha='center', fontsize=12)
plt.title('High Risk vs. Low Risk', fontsize=18)
plt.xlabel('Risk', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)

# Set main title
plt.suptitle('Amplification and Hotspot Mutations of AR', fontsize=20, fontweight='bold')

plt.tight_layout()

# Save the plot as PNG
plt.savefig("C:/Users/hayae/Desktop/AR_high_volume_low_volume_vs_high_risk_low_risk.png")

plt.show()
```

<strong><h1>Total HRR Gene Variants Across Patients</h1></strong>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a DataFrame
df = pd.read_excel("C:/Users/hayae/Desktop/Onco_test.xlsx", sheet_name="HRR")  # Update sheet name if necessary

# Define the list of genes
genes = ['ARID1A', 'ATM', 'ATR', 'ATRX', 'BARD1', 'BRCA1', 'BRCA2', 'BRIP1', 'CDK12', 'CHEK1',
         'CHEK2', 'FANCA', 'FANCF', 'FANCG', 'FANCI', 'FANCL', 'FANCM', 'HDAC2', 'MLH1', 'MRE11A',
         'MSH2', 'NBN', 'PALB2', 'PALP2', 'PPP2R2A', 'RAD50', 'RAD51', 'RAD51B', 'RAD51C', 'RAD51D',
         'RAD54', 'RAD54L', 'WRN']

# Sum the occurrences of each gene for all patients
gene_counts = df[genes].sum(axis=1)

# Sort patients by the total number of gene variants
sorted_indices = gene_counts.argsort()
sorted_gene_counts = gene_counts.iloc[sorted_indices]
sorted_patients = df.iloc[sorted_indices, 0]

# Define patient IDs to be highlighted
highlighted_LPC = ['PR_016', 'PR_018', 'PR_020', 'PR_021', 'PR_270', 'PR_282', 'PR_283', 'PR_46', 'PR_48', 'PR_49']
highlighted_CRPC = ['PR_001', 'PR_015', 'PR_017', 'PR_122', 'PR_135', 'PR_142', 
                        'PR_189', 'PR_201', 'PR_209', 'PR_245', 'PR_36', 'PR_42', 
                        'PR_54', 'PR_56']

# Plotting
plt.figure(figsize=(20, 10))

# Plotting all patients
plt.bar(sorted_patients, sorted_gene_counts, color='#56B4E9', label='Gene Variants')

# Plotting highlighted patients
for patient_id in highlighted_LPC:
    index = df[df.iloc[:, 0] == patient_id].index[0]
    plt.bar(df.iloc[index, 0], gene_counts[index], color='#E69F00')

    for patient_id in highlighted_CRPC:
        index = df[df.iloc[:, 0] == patient_id].index[0]
        plt.bar(df.iloc[index, 0], gene_counts[index], color="#FF9999")

plt.title('Total HRR Gene Variants Across Patients', fontsize=22, fontweight='bold')
plt.xlabel('Patient ID', fontsize=18)
plt.ylabel('Total Gene Variants', fontsize=18)
plt.xticks(rotation=90, fontsize=16)  # Rotate x-axis labels and increase font size
plt.yticks(fontsize=16)
# Create legend
plt.legend(['HSPC', 'LPC', 'CRPC'], fontsize=14)

plt.tight_layout()

# Save the plot as PNG
plt.savefig("C:/Users/hayae/Desktop/sorted_Total_HRR_gene_variants_per_patient.png")

plt.show()
```
